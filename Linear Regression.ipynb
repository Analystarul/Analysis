### 4.Visualization of data

```python
import seaborn as sns
sns.jointplot(x="Temperature",y="Revenue",data=df)
```

### 5. Splitting the dataset

```python
x = df[["Temperature"]].values #2D array ,Independent variable
y = df["Revenue"].values #1D array ,Dependent Variable
```

```python
x.shape,y.shape
```

```python
x
```

```python
y
```

```python
pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split #function
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=2) # Train --> 80%, test --> 20%
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

### 6. Train the model

```python
from sklearn.linear_model import LinearRegression #class
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

### y = mx+c

m--> slope, c--> y intercept

```python
print("Intercept value : ",regressor.intercept_)
```

```python
print("Slope value(a) : ",regressor.coef_)
```

### y = 21.37x + 46.19

```python
y_train_pred = regressor.predict(x_train)
```

```python
y_train_pred.shape
```

```python
y_train.shape
```

### 7.Visualize the train results

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,color="red")
plt.xlabel("Temperature")
plt.ylabel("Revenue")
plt.title("Relationship between Temperature and Revenue")
plt.show
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,color="red")
plt.plot(x_train,y_train_pred,color="black")
plt.xlabel("Temperature")
plt.ylabel("Revenue")
plt.title("Regression line")
plt.show()
```
